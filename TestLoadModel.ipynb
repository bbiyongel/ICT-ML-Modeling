```python
import keras
import numpy as np

load_path='C:/Taeuk/ICT/HumanPoseDetect/Xception/35-0.2064.hdf5'
```

```python
json_file = open("C:/jupyter/Xception.json")
model_json = json_file.read()
json_file.close()

model = keras.models.model_from_json(model_json)
model.load_weights(load_path)

model.compile(
    loss='categorical_crossentropy',
    optimizer=keras.optimizers.RMSprop(lr=1e-4),
    metrics=['acc']
)
```

```python
model.summary() #
```

```python
import os

base_dir = 'C:/Taeuk/HumanPoseDetectDatasets/validation/'
classes = ['center/', 'danger/', 'empty/', 'left/', 'right/', 'sit/']

CENTER = 0
DANGER = 1
EMPTY = 2
LEFT = 3
RIGHT = 4
SIT = 5

for class_name in classes:
    for file_name in os.listdir(base_dir+class_name):
        if file_name.endswith('.jpg'):
            img = keras.preprocessing.image.load_img(base_dir+class_name+file_name, target_size=(299, 299))
            img_tensor = keras.preprocessing.image.img_to_array(img)
            img_tensor = np.expand_dims(img_tensor, axis=0)
            img_tensor /= 255.
            
            predict_classes = model.predict_classes(img_tensor)
            predict = model.predict(img_tensor)
            
            print(class_name+file_name)
            print(predict_classes)
            print(predict)
            print('------------------')
```
